## Data Cleaning

```python
sorted = df.sort_values(['Inventory_number'], ascending = True)

df_copy1 = sorted.groupby('Inventory_number').first().reset_index()
```

```python
df_copy1.drop(columns = ['_id', 'Bibliography','__v'], inplace = True)
```

```python
df_copy1.to_csv("T3_v2.csv")
```

```python
import os
from os import listdir
from os.path import isfile, join
```

```python
os.getcwd()
os.chdir('./imgs')
mypath = os.getcwd()
mypath
```

```python
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles
```

```python
df_copy2 = df_copy1[:]
df_copy2.info()
```

```python
# print(df_copy2["Inventory_number"])
for i in df_copy2["Inventory_number"]:
    z = str(i + '.jpeg')
    if z not in onlyfiles:
        print(z, "true")
        df_copy2.drop(df_copy2[df_copy2.Inventory_number==(i)].index, inplace = True)
```

```python
All_types = df_copy2['Category'].unique()
All_types
```

```python
types_reduced = []

# listy = All_types[0].split(":")
# listy[0]

for i in All_types:
    type = i.split(":")
    if type[0] not in types_reduced:
        types_reduced.append(type[0])
        

types_reduced
```

```python
all_subdirs = [d for d in os.listdir('.') if os.path.isdir(d)]
all_subdirs
```

```python
import shutil 
for i, row in df_copy2.iterrows():
    row_cat = row['Category'].split(":")
    if row_cat[0][-1] == "?":
        row_cat[0] = row_cat[0][:len(row_cat[0])-2]
    if row_cat[0] not in all_subdirs:
        os.makedirs(row_cat[0])
        all_subdirs.append(row_cat[0])
        shutil.move(str(row["Inventory_number"]+'.jpeg'), os.getcwd()+f'\\{row_cat[0]}') 
    else:
        shutil.move(str(row["Inventory_number"]+'.jpeg'), os.getcwd()+f'\\{row_cat[0]}') 
```
